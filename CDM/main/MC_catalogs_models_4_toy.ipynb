```python
def haloloader(index,data_cat,path=path):
    halo = {" " : " "}
    for i in data_cat:
        halo[i] = np.loadtxt(path+i+"_"+str(index)+".txt")
    return halo

halos_av = haloloader(1,data_cat)
halos_4 = haloloader(4,data_cat)
halos_toy = haloloader(5,data_cat)
```

The new model for the above parameters can be calculated:

```python
def subhalo_catalog_MC(m0_g,rs0_g,rhos0_g,ct0_g,weight_g,mth,r_g=None):
    
    condition = (m0_g>mth)
    m0        = m0_g[condition]
    rs0       = rs0_g[condition]
    rhos0     = rhos0_g[condition]
    ct0       = ct0_g[condition]
    weight    = weight_g[condition]
        
    mu_sh      = np.sum(weight)
    prob       = weight/mu_sh
    subhalo_id = np.arange(np.alen(prob))
    N_sh       = np.random.poisson(mu_sh)
    id_MC      = np.random.choice(subhalo_id,size=N_sh,p=prob)

    m0_MC      = m0[id_MC]
    rs0_MC     = rs0[id_MC]
    rhos0_MC   = rhos0[id_MC]
    ct0_MC     = ct0[id_MC]

    if r_g is None:
        return m0_MC/Msolar, rs0_MC/kpc, rhos0_MC/(Msolar/pc**3), ct0_MC 
    else:
        r = r_g[condition]
        r_MC = r[id_MC]
        return m0_MC/Msolar, rs0_MC/kpc, rhos0_MC/(Msolar/pc**3), ct0_MC , r_MC
```

```python
m_min = 1e7
["m0","ct0","rs0","rhos0","r","weight","survive","rs_a","rhos_a"]
m0_toy, rs0_toy, rhos0_toy, ct0_toy , r_toy = subhalo_catalog_MC(halos_toy['m0'],halos_toy['rs0'],
                halos_toy['rhos0'],halos_4['ct0'],halos_4['weight'],m_min*Msolar,r_g=halos_4['r'])

m0_ns, rs0_ns, rhos0_ns, ct0_ns = subhalo_catalog_MC(halos_4['m0'],halos_4['rs0'],halos_4['rhos0'],
                                                    halos_4['ct0'],halos_4['weight'],m_min*Msolar)

m0_4, rs0_4, rhos0_4, ct0_4 , r_4 = subhalo_catalog_MC(halos_4['m0'],halos_4['rs0'],halos_4['rhos0'],
                                                    halos_4['ct0'],halos_4['weight'],m_min*Msolar,r_g=halos_4['r'])
```

```python
def annihilation_boost_factor(rhos0,rs0,ct0,mass,weight,Mhost=M0,redshift=0, evolved=True):    
    fsh = mass_fraction(mass=m0_new,weight=weight_new ,Mhost=M0,evolved=True)
    if evolved:
        Lsh = np.sum(rhos0**2*rs0**3*(1.-1./(1.+ct0)**3)*weight)
    else:
        r200 = (3.*mass/(4.*np.pi*rhocrit(redshift)*200.))**(1./3.)
        c200 = r200/rs0
        Lsh  = np.sum(rhos0**2*rs0**3*(1.-1./(1.+c200)**3)*weight)
    Mhost=Msolar*Mhost
    r200_host = (3.*Mhost/(4.*np.pi*rhocrit(redshift)*200.))**(1./3.)
    c200_host = conc200(Mhost,redshift)
    rs_host   = r200_host/c200_host
    rhos_host = Mhost/(4.*np.pi*rs_host**3*fc(c200_host))
    Lhost0    = rhos_host**2*rs_host**3*(1.-1./(1.+c200_host)**3)
    Bsh = Lsh/Lhost0
    luminosity_ratio = 1.-fsh**2+Bsh
    return Bsh, luminosity_ratio
```
