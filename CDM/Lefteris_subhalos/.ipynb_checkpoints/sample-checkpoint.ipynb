```python
#import numpy as np
#import matplotlib.pyplot as plt
from subhalos import *

plt.rc('font', **{'family':'sans-serif','size':15})
plt.rc('text', usetex=False)
plt.rc('xtick', labelsize=15)
plt.rc('ytick', labelsize=15)
plt.rcParams['xtick.major.size'] = 5
plt.rcParams['xtick.major.width'] = 1
plt.rcParams['xtick.minor.size'] = 2
plt.rcParams['xtick.minor.width'] = 1
plt.rcParams['ytick.major.size'] = 5
plt.rcParams['ytick.major.width'] = 1
plt.rcParams['ytick.minor.size'] = 2
plt.rcParams['ytick.minor.width'] = 1
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
plt.rcParams['xtick.major.pad'] = 8
plt.rcParams['xtick.top'] = True
plt.rcParams['ytick.right'] = True
```

Choose data file for $\sigma(m)$ and $d\ln\sigma(m)/d\ln m$.
This example file "tab_sigma_Sk_kcut1d2.dat" corresponds to the primordial power spectrum just as CDM but with sudden truncation at 10$^2$ Mpc$^{-1}$. 

```python
filename = 'tab_sigma_Sk_kcut1d2.dat'
```

Then, call the class "subhalos" with the filename defined above.

```python
sh = subhalos(filename)
```

The function "rs_rhos_calc" under the "subhalo" class returns list of subhalos with various quantities such as mass, $r_s$, $\rho_s$, etc., both at accretion $z = z_a$ and at present $z = 0$. It also returns "weight" that is effective number of subhalos corresponding to each "subhalo".

```python
ma200, z_a, rs_a, rhos_a, m0, rs0, rhos0, ct0, weight, survive \
    = sh.rs_rhos_calc(1e12,logmamin=5,N_ma=500)
```

Make sure to multiply each quantity by proper units such that all these quantities have a proper dimension.

```python
ma200  *= Msolar
m0     *= Msolar
rs_a   *= kpc
rs0    *= kpc
rhos_a *= Msolar/pc**3
rhos0  *= Msolar/pc**3
```

We can also construct other quantities such as $V_{\rm max}$ and $r_{\rm max}$.

```python
rpeak = 2.163*rs_a
rmax  = 2.163*rs0
Vpeak = np.sqrt(4.*np.pi*G*rhos_a/4.625)*rs_a
Vmax  = np.sqrt(4.*np.pi*G*rhos0/4.625)*rs0
```

Let us now get subhalo mass function as an example. You can just draw histogram of the present mass $m_0$. Make sure to include proper weights.

```python
N,lnm_edges = np.histogram(np.log(m0/Msolar),weights=weight,bins=100)
lnm = (lnm_edges[1:]+lnm_edges[:-1])/2.
dlnm = lnm_edges[1:]-lnm_edges[:-1]

m = np.exp(lnm)
dNdlnm = N/dlnm
dNdm   = dNdlnm/m

plt.loglog(m,m**2*dNdm)
plt.xlim(1e5,1e12)
plt.ylim(1e8,1e11)
plt.xlabel(r'$m$ [$M_{\odot}$]',fontsize=20)
plt.ylabel(r'$m^{2} dN_{\rm sh}/dm$ [$M_{\odot}$]',fontsize=20)
plt.show()
```

```python
N,lnm_edges = np.histogram(np.log(m0/Msolar),weights=weight,bins=100)
lnm = (lnm_edges[1:]+lnm_edges[:-1])/2.
dlnm = lnm_edges[1:]-lnm_edges[:-1]

m = np.exp(lnm)
dNdlnm = N/dlnm
dNdm   = dNdlnm/m

plt.loglog(m,dNdlnm)
plt.xlim(1e5,1e12)
#plt.ylim(1e8,1e11)
plt.xlabel(r'$m$ [$M_{\odot}$]',fontsize=20)
plt.ylabel(r'$dN_{\rm sh}/dlnm$ [$M_{\odot}$]',fontsize=20)
plt.show()
```

Now we get cumulative distribution of $V_{\rm max}$. This is an observable quantity, so one can test warm dark matter hypothesis compared with, e.g., number of observed dwarf galaxies as a function of $V_{\rm max}$.

```python
N,x_edges = np.histogram(Vmax/(km/s),weights=weight,bins=10000)
x = (x_edges[1:]+x_edges[:-1])/2.
Ncum = np.cumsum(N)
Ncum = Ncum[-1]-Ncum
plt.semilogy(x,Ncum)
plt.xlim(0,70)
plt.ylim(1e-1,1e3)
plt.xlabel(r'$V_{\rm max}$ [km/s]',fontsize=20)
plt.ylabel(r'$N_{\rm sh}(>V_{\rm max})$',fontsize=20)
plt.show()
```

In reality, however, not all the subhalos form a satellite galaxy in them. One should impose a "satellite-forming condition" somehow. One very simple prescription is to impose $V_{\rm peak} > 18$ km/s, roughly corresponding to threshold of hydrogen cooling. After doing so, the cumulative distribution looks like as follows.

```python
N,x_edges = np.histogram(Vmax[Vpeak>18*km/s]/(km/s),weights=weight[Vpeak>18*km/s],bins=10000)
x = (x_edges[1:]+x_edges[:-1])/2.
Ncum = np.cumsum(N)
Ncum = Ncum[-1]-Ncum
plt.semilogy(x,Ncum)
plt.xlim(0,70)
plt.ylim(1e-1,1e3)
plt.xlabel(r'$V_{\rm max}$ [km/s]',fontsize=20)
plt.ylabel(r'$N_{\rm sat}(>V_{\rm max})$',fontsize=20)
plt.show()
```
