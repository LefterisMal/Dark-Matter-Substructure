```python
def sat_vth(Vmax,Vpeak,Vpeak_max,weight):
    N,x_edges = np.histogram(Vmax[Vpeak>Vpeak_max*km/s]/(km/s),weights=weight[Vpeak>Vpeak_max*km/s],bins=10000)
    x = (x_edges[1:]+x_edges[:-1])/2.
    Ncum = np.cumsum(N)
    Ncum = Ncum[-1]-Ncum

    return Ncum[0], x, Ncum

def m2dndm(m0,weight,zacc=None,z_min=0,z_max=7):
    
    if (zacc is None)==False:
        m0=m0[(zacc>z_min)*(zacc<=z_max)]
        weight=weight[(zacc>z_min)*(zacc<=z_max)]
        
    N,lnm_edges = np.histogram(np.log(m0/Msolar),weights=weight)#,bins=100)
    lnm = (lnm_edges[1:]+lnm_edges[:-1])/2.
    dlnm = lnm_edges[1:]-lnm_edges[:-1]

    m = np.exp(lnm)
    dNdlnm = N/dlnm

    dNdm   = dNdlnm/m
    

    return m, m**2*dNdm

m_1,m2dndm_1 = m2dndm(m0_1,weight_1)
m_2,m2dndm_2 = m2dndm(m0_2,weight_2)
m_3,m2dndm_3 = m2dndm(m0_3,weight_3)
m_4,m2dndm_4 = m2dndm(m0_4,weight_4)
m_5,m2dndm_5 = m2dndm(m0_5,weight_5)
m_6,m2dndm_6 = m2dndm(m0_6,weight_6)
m_7,m2dndm_7 = m2dndm(m0_7,weight_7)
m_8,m2dndm_8 = m2dndm(m0_8,weight_8)
```

```python
def slope_1(m0,weight):
    N,lnm_edges = np.histogram(np.log(m0/Msolar),weights=weight,bins=30)
    lnm = (lnm_edges[1:]+lnm_edges[:-1])/2.
    dlnm = lnm_edges[1:]-lnm_edges[:-1]
    dlnm_2 = (dlnm[1:]+dlnm[:-1])/2
    m = np.exp(lnm)
    m_2 = (m[1:]+m[:-1])/2.
    dNdm = N/dlnm/m
    dNdm_2= np.log(dNdm[1:])-np.log(dNdm[:-1])
    return m_2,dNdm_2/dlnm_2

m2_1, a_1 = slope_1(m0_1,weight_1)
m2_2, a_2 = slope_1(m0_2,weight_2)
m2_3, a_3 = slope_1(m0_3,weight_3)
m2_4, a_4 = slope_1(m0_4,weight_4)
m2_5, a_5 = slope_1(m0_5,weight_5)
m2_6, a_6 = slope_1(m0_6,weight_6)
m2_7, a_7 = slope_1(m0_7,weight_7)
m2_8, a_8 = slope_1(m0_8,weight_8)

plt.figure(figsize=(7,7))
plt.plot(m2_1, a_1)
plt.plot(m2_2, a_2)
plt.plot(m2_3, a_3)
plt.plot(m2_4, a_4)
plt.plot(m2_5, a_5)
plt.plot(m2_6, a_6)
plt.plot(m2_7, a_7)
plt.plot(m2_8, a_8)
plt.xlim(1e-5,1e12)
plt.ylim(-2.2,-1.6)
plt.xscale('log')
plt.xlabel(r"$m[M\odot]$")
plt.ylabel("a")
plt.legend(['dz=0.008','dz=0.01','dz=0.05','dz=0.1','dz=0.3','dz=0.7','dz=1.2','dz=1.7'],fontsize=10)
plt.title('Slope: $M_{host}=10^{ %i } M_{\odot}$' %np.log10(M0),loc='center')
plt.tight_layout()
#plt.savefig('slope_dz_comp_no_spatial.png')
plt.show()
```

```python
import scipy.integrate as spint
def mass_fraction(mass,weight,Mhost=1e12,evolved=True):
    Mhost=Mhost*Msolar
    if evolved:
        fsh = np.sum(mass*weight)/Mhost
    else:
        fsh = np.sum(mass*weight)/Mhost
    return fsh

def mass_fraction_2(mass,weight,Mhost=M0):
    Mhost=Mhost*Msolar
    N,lnm_edges = np.histogram(np.log(mass/Msolar),weights=weight,bins=100)
    lnm = (lnm_edges[1:]+lnm_edges[:-1])/2.
    dlnm = lnm_edges[1:]-lnm_edges[:-1]
    m = np.exp(lnm)
    dNdm = N/dlnm/m
    
    fsh=spint.simpson(dNdm*m,m)/M0
    return fsh

print(" Weighted Sum Calculation:")
print(" Substructure mass fraction for dz=0.008:",
      mass_fraction(mass=m0_1,weight=weight_1 ,Mhost=M0,evolved=True))


print(" Substructure mass fraction for dz=0.01:",
      mass_fraction(mass=m0_2,weight=weight_2 ,Mhost=M0,evolved=True))

print(" Substructure mass fraction for dz=0.05:",
      mass_fraction(mass=m0_3,weight=weight_3 ,Mhost=M0,evolved=True))

print(" Substructure mass fraction for dz=0.1:",
      mass_fraction(mass=m0_4,weight=weight_4 ,Mhost=M0,evolved=True))

print(" Substructure mass fraction for for dz=0.3:",
      mass_fraction(mass=m0_5,weight=weight_5 ,Mhost=M0,evolved=True))

print(" Substructure mass fraction for dz=0.7:",
      mass_fraction(mass=m0_6,weight=weight_6 ,Mhost=M0,evolved=True))

print(" Substructure mass fraction for for dz=1.2:",
      mass_fraction(mass=m0_7,weight=weight_7 ,Mhost=M0,evolved=True))

print(" Substructure mass fraction for for dz=1.7:",
      mass_fraction(mass=m0_8,weight=weight_8 ,Mhost=M0,evolved=True))


print("\n _________________________________________________________________________________ \n")
print(" Integral Calculation:")
print(" Substructure mass fraction for dz=0.008:",
      mass_fraction_2(mass=m0_1,weight=weight_1 ,Mhost=M0))


print(" Substructure mass fraction for dz=0.01:",
      mass_fraction_2(mass=m0_2,weight=weight_2 ,Mhost=M0))

print(" Substructure mass fraction for dz=0.05:",
      mass_fraction_2(mass=m0_3,weight=weight_3 ,Mhost=M0))

print(" Substructure mass fraction for dz=0.1:",
      mass_fraction_2(mass=m0_4,weight=weight_4 ,Mhost=M0))

print(" Substructure mass fraction for dz=0.3:",
      mass_fraction_2(mass=m0_5,weight=weight_5 ,Mhost=M0))

print(" Substructure mass fraction for dz=0.7:",
      mass_fraction_2(mass=m0_6,weight=weight_6 ,Mhost=M0))

print(" Substructure mass fraction for dz=1.2:",
      mass_fraction_2(mass=m0_7,weight=weight_7 ,Mhost=M0))

print(" Substructure mass fraction for dz=1.7:",
      mass_fraction_2(mass=m0_8,weight=weight_8 ,Mhost=M0))
```

```python
import scipy.optimize as spopt
import scipy.stats as sps
def fiting(function,x,y,p0):
    ml_cfpars, ml_cfcovar = spopt.curve_fit(function,x,y,p0)
    print("The covariance matrix is: ",ml_cfcovar)
    model_ens = np.linspace(0,1.8,1000)

    fig, (ax1, ax2) = plt.subplots(2,1, figsize=(8,6),sharex=True,gridspec_kw={'height_ratios':[2,1]})
    fig.subplots_adjust(hspace=0)
    y_fit=function(model_ens,*ml_cfpars)
    ax1.set_title(r"Substructure mass fraction for $M_{host}=10^{12} \cdot M_{solar}$", fontsize=12)
    ax1.plot(model_ens, y_fit, lw=2)
    ax1.scatter(x, y)
    ax1.set_xscale('log')
    ax1.set_ylabel(r"$\frac{m_{sh,total}}{M_{host}}$", fontsize=16)
    
    ax2.scatter(x,function(x,*ml_cfpars)/y)
    ax2.set_ylabel("model/data", fontsize=12)
    ax2.set_xlabel("dz (redshift step)",fontsize=12)
    ax2.axhline(1.0, color='r', linestyle='dotted', lw=2)
    ax2.set_ylim(0.8,1.2)
    plt.show()

    print("Model's prediction of mass fraction for dz=0 value",function(0,*ml_cfpars))
    return ml_cfpars
```

```python
def pol_2(x,*par):
    return par[2]*x**2+par[1]*x+par[0]

def pol_3(x,*par):
    return par[3]*x**3+par[2]*x**2+par[1]*x+par[0]
```
