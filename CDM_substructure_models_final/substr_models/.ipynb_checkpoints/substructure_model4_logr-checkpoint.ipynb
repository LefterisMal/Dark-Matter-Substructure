# Spatial result of the updated model and simulation

Here I define the functions that are the result of Shunichi 's work on radial linear probabability distribution,with the function Simulation_2(x) being more accurate.The functions Simulation_1(x),Simulation_2(x) can be found at slide 23 and 24 at Shunichi's slides. The Simulation_1(x) is noted as (4.1) and Simulation_2(x) as (4.2) by the respected equations in my notion page.

```python
#Radial distribution from simulation's results:

#First results by fitting:
def Simulation_1(x):
    c=0.821
    S=1.519
    return S*(c*x**2)/(c*x**2+(1-c)*(1-x)**2)

#Second results (and better one) by fitting:
def Simulation_2(x):
    a=2.2
    c=0.606
    S=9.839
    return S*x**2/(1+(x/c)**2)**(3/a)

#Function for selecting subhalos' mass within a specific mass range,the variable m0 is an array:

def mass_selection(m_min,m_max,m0):
    mass_range_low=m0>(m_min*Msolar)
    mass_range_high=m0<(m_max*Msolar)
    return mass_range_low*mass_range_high
```

# Below some important functions that I use in my model are presented and could lead to the solution of the disagreement between the simulation and my model

Below I show the functions that I am using in my model for the dm/dt and radial distribution (initial radial distribution) before tidal stripping.The dm/dt model 2 is presented at the slide 13 in Shunichi's slides and model 3 at the slide 17 (the number of the dm/dt models are the ones that I use in the function mdot_r below).The dm/dt model 3 was calculated by data in the range $\frac{distance}{R_{vir}} \in [0.2,1]$ and for $\frac{distance}{R_{vir}} \in [0,0.2]$ the extrapolation of this model was used. 

```python
def mdot_r(r,z,model,M0=M0):
    if model==2:
        return 10**(0.2-1.75*0.3*r)
    elif model==3:
        a=1.34
        b=1.39
        c=2.87
        eq=a-b*(1+(1-r)**c)**(-1/c)
        return 10**eq
    elif model==4:
        con=c_from_Mvir(M0,z)
        if hasattr(r, "__len__")==False:
            if r>0.2:
                a=1.34
                b=1.39
                c=2.87
                eq=a-b*(1+(1-r)**c)**(-1/c)
                return 10**eq
            else:
                return np.log((np.log(1+con)-con/(1+con))/(np.log(1+r*con)-(con*r)/(1+con*r)))
        else:
            mdot_c=np.zeros(len(r))
            n=0
            for i in r:
                if i>0.2:
                    a=1.34
                    b=1.39
                    c=2.87
                    eq=a-b*(1+(1-i)**c)**(-1/c)
                    mdot_c[n]=10**eq
                else:
                    mdot_c[n]=np.log((np.log(1+con)-con/(1+con))/(np.log(1+i*con)-(con*i)/(1+con*i)))
                n+=1
            return mdot_c
    elif model==5:
        c=c_from_Mvir(Mvir=M0*Msolar,z=z)
        return np.log((np.log(1+c)-c/(1+c))/(np.log(1+r*c)-(c*r)/(1+c*r)))
    else:
        return 1
    
    
def radial_weights(r,M=M0):
    c=c_from_Mvir(Mvir=M*Msolar,z=7)
    M_all=np.log(1+c)-c/(1+c)
    Prob=(r*c**2)/((1+c*r)**2)
    return Prob/M_all
```

```python
r=np.linspace(0.01,1,1000)
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(12,5))
fig.subplots_adjust(wspace=0.3)

for ax in (ax1,ax2):
    ax.tick_params(labelsize=12)
    ax.set_xlabel(r'$\frac{distance}{R_{vir}}$', fontsize=12)
    ax.set_ylabel(r'$\frac{dm}{dt} [\overline{\frac{dm}{dt}}]$', fontsize=12)
    ax.plot(r,mdot_r(r,0,2,M0=M0))
    ax.plot(r,mdot_r(r,0,3,M0=M0))
    ax.plot(r,mdot_r(r,0,5,M0=M0))
    ax.plot(r,mdot_r(r,7,4,M0=M0))
    ax.legend([r'$\frac{dm}{dt}$ model:2',r'$\frac{dm}{dt}$ model:3',
               r'$\frac{dm}{dt}$ model:toy model',r'$\frac{dm}{dt}$ model:4'],fontsize=10,loc='best')
    ax.set_title(r'$M_{host}=10^{12}M_{\odot}$,$\frac{dm}{dt}$ in units of $[\overline{\frac{dm}{dt}}]$'
                 , fontsize=10)
ax2.set_ylim(1e-1,8*1e0)
ax2.set_xscale("log")
ax2.set_yscale("log")
plt.tight_layout()
```
